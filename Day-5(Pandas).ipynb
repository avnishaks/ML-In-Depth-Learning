# using tuple list

```python
tuple_list=[
    ('Monday','Rain',10,'1/1/22'),
    ('Tuesday','Snow',20,'2/2/22'),
    ('Wednesday','Ice',30,'3/3/22')
]
```

```python
# we have to provide the column name in this
tuple_data=pan.DataFrame(tuple_list,columns=["day","weather","Amount","year"])
```

```python
tuple_data
```
