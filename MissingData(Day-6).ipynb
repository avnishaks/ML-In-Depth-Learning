# Topics covered:-
1. Fillna to Fill missing values using different ways.

# Fill_na

```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv')
```

```python
df
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv',parse_dates=["day"])
```

```python
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':5,
    'windspeed':2,
    'event':'no event'
})
new_df
```

```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
horizontal_back_data=df.fillna(method='bfill',axis='columns')
```

```python
horizontal_back_data
```

```python
horizontal_foreward_data=df.fillna(method='bfill',axis='columns')
```

```python
horizontal_foreward_data
```

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```

```python
new_df=df.fillna(method='ffill',limit=3)
new_df
```

# Replace

```python
import pandas as pd
import numpy as np
```

```python
new_df=df.replace(np.NaN,-9999)
new_df
```

```python
new_df=df.fillna({
    'event':'no event'
})
new_df
```

```python
new_df=new_df.fillna({
    'temperature':-9999,
    'windspeed':-8888
})
new_df
```

```python
new_df=new_df.replace([-9999, -8888],np.NaN)
new_df
```

```python
new_df=new_df.replace({
    'event':'no event',
},np.NaN)
new_df
```

```python
new_df=new_df.replace({
    'event':np.NaN,
},'LOVE')
```

```python
new_df
```

```python
new_df=new_df.fillna({
    'temperature':'32 F',
    'windspeed': '6 mph'
})
```

```python
new_df
```

```python
new_df=new_df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```
